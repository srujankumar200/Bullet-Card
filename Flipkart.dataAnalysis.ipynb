```python
categories=['Primar_category','Secondary_category','Tertiary_category','Quadernary_category']
```

```python
def categorical_values():
    for num in [0,1,2,3]:
        if(num==0):
            data[categories[num]]=data['product_category_tree'].apply(lambda x:x.split('>>')[num])
        if(num==1):
            data[categories[num]]=data['product_category_tree'].apply(lambda x:x.split('>>')[num:num+1])
        if(num==2):
            data[categories[num]]=data['product_category_tree'].apply(lambda x:x.split('>>')[num:num+1])
        if(num==3):
            data[categories[num]]=data['product_category_tree'].apply(lambda x:x.split('>>')[num:num+1])
categorical_values()
```

#creation of new columns based on the above mentioned categories

```python
data.head()
```

```python
data['Primar_category'] = data['Primar_category'].apply(lambda x: x.split('["')[0])
```

```python
for i in [1,2,3]:
    data[categories[i]] = data[categories[i]].apply(lambda x: str(x))
    data[categories[i]] = data[categories[i]].apply(lambda x: x[2:-2])
```
